## Lab 3 - Bài 1
Viết chương trình thực hiện các thao tác sau với một dãy số nguyên. 
✓ Sắp xếp theo thứ tự tăng dần và xuất dãy số nguyên ra màn hình. 
✓ Xuất phần tử có giá trị nhỏ nhất ra màn hình 
✓ Tính và xuất ra màn hình trung bình cộng các phần tử chia hết cho 3 

```python
# Code lab 3 bài 1 ở đây
# Nhập dãy số nguyên từ người dùng
dayso = list(map(int, input("Nhập dãy số nguyên, cách nhau bởi dấu cách: ").split()))

#1. Sắp xếp theo thứ tự tăng dần
daysotangdan = sorted(dayso)
print("Dãy số sau khi sắp xếp tăng dần:", daysotangdan)

#2. Xuất phần từ có giá trị nhở nhất
giatrimin = min(dayso)
print("Phần tử có giá trị nhỏ nhất trong dãy số là:", giatrimin)

#3 .Tính và xuất ra màn hình trung bình cộng các phần tử chia hết cho 3
cacsochiahetcho3 = [x for x in dayso if x % 3 == 0]
if cacsochiahetcho3:
    trungbinhcong = sum(cacsochiahetcho3) / len(cacsochiahetcho3)
    print("Trung bình cộng các phần tử chia hết cho 3 là:", trungbinhcong)

# 4. Sắp xếp theo thứ tự giảm dần
daysogiamdan = sorted(dayso, reverse=True)
print("Dãy số sau khi sắp xếp giảm dần:", daysogiamdan)
```

## Lab 3 - Bài 2
Viết chương trình nhập dữ liệu của một sinh viên. Thông tin của sinh viên bao 
gồm: họ tên và điểm.

```python
# Code lab 3  bài 2 ở đây
# Nhập thông tin của  sinh viên
danhsachsinhvien = []
soluongsinhvien = int(input("Nhập số lượng sinh viên: "))
for i in range(soluongsinhvien):
    HoTen = input(f"Nhập họ tên sinh viên thứ {i+1}: ")
    Diem = float(input(f"Nhập điểm của {HoTen}"))

# Xếp loại
    if Diem < 5 :
        Xeploai = "Yếu"
    elif Diem >=5 and Diem < 6.5 :
        Xeploai = "Trung bình"
    elif Diem >= 6.5 and Diem < 7.5 :
        Xeploai = "Khá"
    elif Diem >= 7.5 and Diem < 9 :
        Xeploai = "Giỏi"
    else:
        Xeploai = "Xuất sắc"
    danhsachsinhvien.append((HoTen, Diem, Xeploai))

# Xuất thông tin sinh viên
print("\nDanh sách sinh viên:")
for sv in danhsachsinhvien:
    print(f"Họ tên: {sv[0]}, Điểm: {sv[1]}, Xếp loại: {sv[2]}")
```

## Lab 3 - Bài 3
Viết chương trình nhập một số nguyên từ bàn phím và cho biết số đó có phải là số 
nguyên tố hay không (số nguyên tố là số chỉ chia hết cho 1 và chính nó). 

```python
# Code lab 3  bài 3 ở đây
songuyen = int(input("Nhập một số nguyên: "))
if songuyen <=1:
    print(f"{songuyen} không phải là số nguyên tố.")
else:
    LaSoNguyenTo = True
    for i in range(2, int(songuyen**0.5) + 1):
        if songuyen % i == 0:
            LaSoNguyenTo = False
            break
    if LaSoNguyenTo:
        print(f"{songuyen} là số nguyên tố.")
    else:
        print(f"{songuyen} không phải là số nguyên tố.")
```

## Lab 3 - Bài 5: bài tập làm thêm

```python
# code lab 3  bài làm thêm ở đây 
import math
import time

LichSu = []

while True:
    print("\n--- MÁY TÍNH BỎ TÚI ---")
    print("1. Phép tính cơ bản")
    print("2. Lũy Thừa")
    print("3. Căn Bậc Hai")
    print("4. Hàm Lượng Giác")
    print("5. Logarit")
    print("6. Giải phương trình bậc nhất")
    print("7. Giải phương trình bậc hai")
    print("8. Xem lịch sử tính toán")
    print("9. Xem thời gian hiện tại")
    print("0. Thoát")
    ChonChucNang = float(input("Chọn một tùy chọn (1-9): "))
    if ChonChucNang == 1:
        ToanHang1 = float(input("Nhập số thứ nhất: "))
        ToanHang2 = float(input("Nhập số thứ hai: "))
        PhepToan = input("Chọn phép toán (+, -, *, /): ")
        if PhepToan == '+':
            KetQua = ToanHang1 + ToanHang2
        elif PhepToan == '-':
            KetQua = ToanHang1 - ToanHang2
        elif PhepToan == '*':
            KetQua = ToanHang1 * ToanHang2
        elif PhepToan == '/':
            if ToanHang2 != 0:
                KetQua = ToanHang1 / ToanHang2
            else:
                print("Lỗi: Không thể chia cho 0.")
                continue
        else:
            print("Phép toán không hợp lệ.")
        print ("Ket qua phep toan là:", KetQua)
        LichSu.append(f"{ToanHang1} {PhepToan} {ToanHang2} = {KetQua}")
    elif ChonChucNang == 2:
        CoSo = float(input("Nhập cơ số: "))
        SoMu = float(input("Nhập số mũ: "))
        KetQua = CoSo ** SoMu
        print ("Ket qua phep toan là:", KetQua)
        LichSu.append(f"{CoSo}^{SoMu} = {KetQua}")
    elif ChonChucNang == 3:
        So = float(input("Nhập số cần tính căn bậc hai: "))
        if So >= 0:
            KetQua = math.sqrt(So)
            print ("Ket qua phep toan là:", KetQua)
            LichSu.append(f"√{So} = {KetQua}")
        else:
            print("Lỗi: Không thể tính căn bậc hai của số âm.")
    elif ChonChucNang == 4:
        hamluonggiac = input("Chọn hàm lượng giác (sin, cos, tan)")
        goc = float(input("Nhập góc (độ): "))
        if hamluonggiac == "sin":
            KetQua = math.sin(math.radians(goc))
        elif hamluonggiac == "cos":
            KetQua = math.cos(math.radians(goc))
        elif hamluonggiac == "tan":
            KetQua = math.tan(math.radians(goc))
        elif hamluonggiac == "cot":
            if math.radians(goc) !=0 :
                KetQua = 1 / math.tan(math.radians(goc))
            else:
                print("Lỗi: Không thể tính cot của góc 0 độ.")
                continue
        else:
            print("Hàm lượng giác không hợp lệ.")
            continue
        print ("Ket qua phep toan là:", KetQua)
        LichSu.append(f"{hamluonggiac}({goc}) = {KetQua}")
    elif ChonChucNang == 5:
        LoaiLog = input("Chọn loại logarit (log, ln, log10): ")
        So = float(input("Nhập số cần tính logarit: "))
        if LoaiLog == "log10":
            if So > 0:
                KetQua = math.log10(So)
            else:
                print("Lỗi: Không thể tính logarit của số âm hoặc bằng 0.")
                continue
        elif LoaiLog == "ln":
            if So > 0:
                KetQua = math.log(So)
            else:
                print("Lỗi: Không thể tính logarit của số âm hoặc bằng 0.")
                continue
        elif LoaiLog == "log":  
            CoSo = float(input("Nhập cơ số (hơn 0 và khác 1): "))
            if So > 0 and CoSo > 0 and CoSo != 1:
                KetQua = math.log(So, CoSo)
            else:
                print("Lỗi: Không thể tính logarit với điều kiện đã cho.")
                continue
        else:
            print("Loại logarit không hợp lệ.")
            continue
        print ("Ket qua phep toan là:", KetQua)
        LichSu.append(f"{LoaiLog}({So}) = {KetQua}")
    elif ChonChucNang == 6:
        a = float(input("Nhập hệ số a (khác 0): "))
        if a == 0:
            print("Lỗi: Hệ số 'a' phải khác 0.")
            continue
        b = float(input("Nhập hệ số b: "))
        KetQua = -b / a
        print ("Ket qua phep toan là:", KetQua)
        LichSu.append(f"{a}x + {b} = 0 => x = {KetQua}")
    elif ChonChucNang == 7:
        a = float(input("Nhập hệ số a (khác 0): "))
        if a == 0:
            print("Lỗi: Hệ số 'a' phải khác 0.")
            continue
        b = float(input("Nhập hệ số b: "))
        c = float(input("Nhập hệ số c: "))
        Delta = b**2 - 4*a*c
        if Delta > 0:
            x1 = (-b + math.sqrt(Delta)) / (2*a)
            x2 = (-b - math.sqrt(Delta)) / (2*a)
            KetQua = (x1, x2)
            print ("Ket qua phep toan là:", KetQua)
            LichSu.append(f"{a}x^2 + {b}x + {c} = 0 => x1 = {x1}, x2 = {x2}")
        elif Delta == 0:
            x = -b / (2*a)
            KetQua = (x,)
            print ("Ket qua phep toan là:", KetQua)
            LichSu.append(f"{a}x^2 + {b}x + {c} = 0 => x = {x}")
        else:
            print("Phương trình vô nghiệm.")
            LichSu.append(f"{a}x^2 + {b}x + {c} = 0 => Vô nghiệm")
    elif ChonChucNang == 8:
        if LichSu:
            print("\n--- Lịch sử tính toán ---")
            for entry in LichSu:
                print(entry)
        else:
            print("Chưa có lịch sử tính toán.")
    elif ChonChucNang == 9:
        print("Thời gian hiện tại:", time.strftime("%H:$M:%S %d/%m/%Y"))
        break
    elif ChonChucNang == 0:
        print("Thoát chương trình")
        break
    else:
        print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
    
    
```
